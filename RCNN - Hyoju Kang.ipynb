# R-CNN (Region-Based Convolutional Neural Networks)

- Used for computer vision and specifically object detection
- A segmentation algorithm (classify 2,000 region proposals per image)
- Compares nearby pixels with similar color, light, etc.. to do selective search and
  pass to CNN network

- Output labels and Bounding boxes

- After Selected cropped pictures are sent to CNN, the neural networks determines what class they are

- Cons: 
       1. all bounding box should be passed by one at a time -> causes slow functioning
       2. cannot be implemented real time as it takes around 47 secs for each test image
       3. Selective search algorithm is a fixed algorithms -> no learning happening at the stage

- typically used for detecting and localizing objets with well-defined boundaries, such as cars, bicycles, and pedestrians. It is challenging for basic R-CNN to detect the lane (especially when lanes have low contrast or are occluded by other objects)
